```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load
file_to_load = "Resources/cities.csv"

# Read Cities File and store into Pandas data frame
cities_data = pd.read_csv(file_to_load)
cities_data
```

```python
clean_cities_data = cities_data.set_index("City_ID")
clean_cities_data
```

```python
clean_cities_data["Cityname"] = clean_cities_data.City.str.title()
clean_cities_data
```

```python
final_cities_data = clean_cities_data[["Cityname","Cloudiness","Country","Date","Humidity","Lat","Lng","Max Temp","Wind Speed"]]
final_cities_data
```

```python
final_cities_data.to_html("cities.html")
```
